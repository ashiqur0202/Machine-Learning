# Linear Regression Algorithm

```python
# Importing libraries
import pandas as pd
```

```python
df = pd.read_csv("data/Height-Weight Data.csv")
```

```python
df.head()
```

```python
x = df[["Height"]]
x.head()
```

```python
y=df[["Weight"]]
y.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Height'], df['Weight'], color='blue')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height-Weight')
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=.75)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
a = lr.intercept_
a
```

```python
b = lr.coef_
b
```

```python
pred = lr.predict(xtest)
pred
```

```python
lr.score(xtest, ytest)
```

```python
df[["New Weight"]] = lr.predict(x)
df.head()
```

```python
plt.scatter(df['Height'], df['New Weight'], color='blue')
plt.xlabel('Height')
plt.ylabel('New Weight')
plt.title('Height-Weight')
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(ytest, pred)
mse
```
